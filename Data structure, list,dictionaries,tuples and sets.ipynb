```python
def add (a,b):
    result = a + b 
    print(result)

def sub (a,b):
    result = a - b
    print(result)

def mult (a,b):
    result = a * b
    print(result)

def div (a,b):
    result = a / b
    print(result)
    
```

```python
options = input("enter your options here:")
options_1 = input("options == A or options == a: "), input("options == S or options == s: "), input("options == M or options == m: "), input("options == D or options == d:")

if options == "A" or options == "a":
    a = int(input("first value"))
    b = int(input("second value"))
    add(a,b)
elif options == "S" or options == "s":
    a = int(input("first value"))
    b = int(input("second value"))
    sub(a,b)
elif options == "M" or options == "m":
    a = int(input("first value"))
    b = int(input("second value"))
    mult(a,b)
elif options == "D" or options == "d":
    a = int(input("first value"))
    b = int(input("second value"))
    div(a,b)
```
